##Linear regression model and implementation


```python
#importing necessary libraries
import pandas as pd
from sklearn import linear_model
```

```python
#loading dataset
df = pd.read_csv('/content/archive.zip')
df.head(10)
```

```python
#renaming (US$) col
df.rename(columns={'per capita income (US$)':'income'}, inplace =True)
df.tail()
```

```python
#creating a linear regression object
reg = linear_model.LinearRegression()
```

```python
#train dataset
reg.fit(df[['year']],df.income)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[2020]])
```

```python
#plotting a scatter diagram; indicatiom that its a linear graph
import matplotlib.pyplot as plt
plt.scatter(df['year'], df['income'], color = 'red')
plt.xlabel('year')
plt.ylabel('per capital income(US$)')
plt.title('a linear graph of per capital income(US$) vs year')

```
