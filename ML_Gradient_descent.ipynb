## Exercise on Gradient Descent, finding m and b value for a particular dataset


```python
#importing libraires
import numpy as np
import pandas as pd
#loading dataset
my_data = pd.read_csv('/content/test_scores.csv')
my_data
```

```python
def gradient_descent (x,y):
  m_curr = b_curr = 0
  iterations = 100
  n = len(x)
  Learning_rate = 0.0004



  for ch in range(iterations):
   y_predicted = m_curr *x + b_curr
   cost = (1/n)*sum([ch**2 for ch in(y-y_predicted)])
   # partial derivative of m_curr
   md = -(2/n)* sum (x*(y-y_predicted))
   # partial derivative of b_curr
   bd = -(2/n)* sum(y-y_predicted)
   # find the next m and b value using a learning rate
   m_curr =  m_curr - Learning_rate * md
   b_curr =  b_curr - Learning_rate * bd

   print('m{},b{},c{},i{}'.format({m_curr}, {b_curr}, {cost}, {iterations}))


x = my_data['math']
#xx = (np.array(x) for i in x)
y = my_data['cs']

gradient_descent (x,y)
```
